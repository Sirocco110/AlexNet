```python
import tensorflow as tf
import keras
from keras import backend as K
from keras.datasets import cifar10

import glob
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tensorflow.python.client import device_lib
device_lib.list_local_devices()
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print("trainデータ")
print(x_train.shape)
print(y_train.shape)
print("testデータ")
print(x_test.shape)
print(y_test.shape)
```

```python
train_classes = [[] for _ in range(10)]
for i in range(len(y_train)):
    train_classes[y_train[i][0]].append(i) 

test_classes = [[] for _ in range(10)]
for i in range(len(y_test)):
    test_classes[y_test[i][0]].append(i) 
```

```python
plt.figure(figsize=(20,20))

for i in range(10):
    for j in range(10):
        plt.subplot(10,10,(10*i)+(j+1))
        plt.axis("off")
        plt.imshow(x_train[train_classes[i][j]])
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
```

```python
input_shape = x_train.shape[1:]
print('入力シェープ = ', input_shape)

num_classes = 10

#AlexNet
model = Sequential()

# Conv1
model.add(Conv2D(96, (3,3),activation='relu', padding='same',
                 input_shape=input_shape))
# Pool1
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(BatchNormalization())


# Conv2
model.add(Conv2D(256, (5,5), activation='relu', padding='same'))
# Pool2
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())


# Conv3
model.add(Conv2D(384, (3,3), activation='relu', padding='same'))

# Conv4
model.add(Conv2D(384, (3,3), activation='relu', padding='same'))

# Conv5
model.add(Conv2D(256, (3,3), activation='relu', padding='same'))
# Pool5
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
# fc6
model.add(Dense(4096, activation='relu'))
# fc7
model.add(Dense(4096, activation='relu'))
# fc8
model.add(Dense(num_classes, activation='softmax'))

model.summary()
```

```python
from keras import optimizers
sgd = optimizers.SGD(lr=0.01)

model.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
batch_size = 128 
epochs = 100 
image_size = 32
channel = 3 

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
# fit関数の定義
history = model.fit (x_train, y_train, # 訓練データと正解ラベルデータ
                     batch_size=batch_size, # バッチサイズ
                     epochs=epochs, # エポック数
                     validation_data=(x_test, y_test)) # テストデータ
```

```python
pred = model.predict(x_test)
pred_classes = np.argmax(pred,axis=1)
```

```python
coll = 0

for i in range(len(y_test)):
    p = pred_classes[i]
    t = y_test[i][0]
    
    if p == t:
        coll += 1

print("正解数 : ",coll)
print("acc : ",coll/len(y_test))
```
